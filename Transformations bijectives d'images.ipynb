## 3 - Image mystere 

```python
mystere= path_module+'/imagemystere.png'
img4= Image.open(mystere)
img4
```

L’image ci-dessus a été obtenu en transformant une image successivement par :
- une transformation du boulanger
- 10000 transformations du boustrophédon consécutives
- 9868756755876 transformations concentrique consécutives
- 5 transfromations du photo-maton consécutives
- et finalement 765765764675587676 transformations concentriques.

Notre objectif est de retrouver l'image initiale.

**D'après le cours, le nombre de périodes pour chaque fonction est :**
- Symétrie : 2
- Boulanger :17
- Photomaton : 8
- Boustrophédon : 65536
- Concentrique : 260538918857747541744672011746506021097973820451713310155553963930295146498906764615494790073109501333431500 

Pour décoder l'image mystère nous allons reprendre les transformations de la dernière à la première, en effectuant le nombre de transformations restantes pour revenir à l'image avant transformation.

```python
img=(TB.transformation_concentrique(mystere,(260538918857747541744672011746506021097973820451713310155553963930295146498906764615494790073109501333431500-765765764675587676)))
img.save(path_module+'/imagemys.png')
img=(TB.transformation_du_photomaton(path_module+'/imagemys.png',(8-5)))
img.save(path_module+'/imagemys.png')
img=(TB.transformation_concentrique(path_module+'/image mys.png',(260538918857747541744672011746506021097973820451713310155553963930295146498906764615494790073109501333431500-9868756755876)))
img.save(path_module+'/imagemys.png')
img=(TB.transformation_boustrophédon(path_module+'/image mys.png',(65536-10000)))
img.save(path_module+'/imagemys.png')
img=(TB.transformation_du_boulanger(path_module+'/image mys.png',(17-1)))

img
```
